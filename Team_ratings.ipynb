```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
stats = pd.read_csv("stats_with_ratings.csv")
finals = ["GF", "PF", "QF", "EF", "SF"]
stats = stats[~stats.Round.isin(finals)]
stats["Round"] = stats["Round"].apply(lambda x: int(x[1:]))
stats = stats.sort_values(["Team", "Season", "Round"])
stats
```

```python
# Getting team ratings for entire seasons, by summing all player ratings for each game

seasons = range(2012, 2019)
teams = {i for i in stats["Team"]}

team_ratings_by_season = dict()

for s in seasons:
    team_ratings = {i: 0 for i in teams}
    season = stats[stats.Season == s]
    split_by_team = season.groupby("Team").sum()
    
    for index, row in split_by_team.iterrows():
        team_ratings[index] = row["Ratings"]
    
    team_ratings_by_season[s] = team_ratings
```

```python
for i in team_ratings_by_season.keys():
    team_ratings_by_season[i] = {k: v for k, v in sorted(team_ratings_by_season[i].items(), \
                                                         key=lambda item: item[1], reverse = True)}
```
